```python
from karyo.cfg import num_classes
from karyo.data import DataModule, DataConfig
from karyo.models import build_model
from karyo.trainer import TrainConfig, Trainer

data_conf = DataConfig()
data = DataModule(data_conf)

train_conf = TrainConfig(num_epochs=40, batch_size=data_conf.batch_size)
model_B = build_model("maskrcnn_r50_fpn", num_classes).to(train_conf.device)

train_loader_B, val_loader_B = data.make_loaders_b_only()
trainer_B = Trainer(model_B, train_conf)
history_B = trainer_B.run(train_loader_B, val_loader_B, experiment_name="maskrcnn_B")
```

```python
model_AB = build_model("maskrcnn_r50_fpn", num_classes).to(train_conf.device)
train_loader_AB, val_loader_AB = data.make_loaders_a_plus_b()
trainer_AB = Trainer(model_AB, train_conf)
history_AB = trainer_AB.run(train_loader_AB, val_loader_AB, experiment_name="maskrcnn_AB")
```
